### Task 2.2: Delete the useless columns

### Task 2.3: Aggregating the rows by the country

```python
dfa=df.groupby("Country/Region").sum() #dfa=data frame aggregated
```

### task 3.1: caculating the first derivative of the curve

### task 3.2: find maxmimum infection rate for China, Italy, Spain and Nigeria


```python
dfa.loc['China'].diff().max()
```

```python
dfa.loc['Nigeria'].diff().max()
```

```python
dfa.loc['Italy'].diff().max()
```

```python
dfa.loc['Spain'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(dfa.index)
mir=[]# max infection rate
for c in countries:
    mir.append(dfa.loc[c].diff().max())
mir
```

```python
dfa["Max infection rate"]=mir
```

```python
dfa.head(10)
```

### Task 3.4: create a new dataframe with only needed column 

```python
nd=pd.DataFrame(dfa["Max infection rate"])#new data
nd.head(10)
```
